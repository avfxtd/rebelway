```python
import seaborn as sns
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
sns.heatmap(df.pivot_table(values="Age",index="Fare",columns="Pclass"))
```

```python
threshold = 300
outliers = df[df['Fare'] >= threshold].index
df.drop(outliers, inplace=True)
```

```python
sns.set_theme(style="ticks")
sns.lmplot(data=df,x="Age",y="Fare",hue="Sex")
```

```python
sns.set_theme(style="white")
sns.relplot(x="Age", y="Fare", hue="Sex", sizes=(40, 400), alpha=.5, palette="muted", height=6, data=df)
```
