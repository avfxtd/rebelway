```python
column_objects = df.drop('Selling_Price',axis=1).select_dtypes(include=['object']).columns
for column_name in column_objects:
    vals = df[column_name].values
    vals = list(set(vals))
    vals_map = {}
    vals_map = dict((j, i) for i, j in enumerate(vals))
    print(vals_map)
    df[column_name] = df[column_name].replace(vals_map)
```

```python
df.head()
```

Single Decision Tree

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X = df.drop('Selling_Price', axis=1)
y = df['Selling_Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=55)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train,y_train)
```

```python
tree_pred = decision_tree.predict(X_test)
```

```python
print(confusion_matrix(y_test,tree_pred))
print("------------------")
print(classification_report(y_test,tree_pred))
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest_model = RandomForestClassifier()
random_forest_model.fit(X_train,y_train)
```

```python
random_preds = random_forest_model.predict(X_test)
```

```python
print(confusion_matrix(y_test,random_preds))
print("------------------")
print(classification_report(y_test,random_preds))
```

Random Forest RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
random_forest_model = RandomForestRegressor()
random_forest_model.fit(X_train,y_train)
```

```python
random_preds = random_forest_model.predict(X_test)
```

```python
plt.scatter(y_test,random_preds)
plt.xlabel("Y test")
plt.ylabel("Predicted Y")
```

```python
print("MAE: ", metrics.mean_absolute_error(y_test,random_preds))
print("MSE: ", metrics.mean_squared_error(y_test,random_preds))
print("RMSE: ", np.sqrt(metrics.mean_squared_error(y_test,random_preds)))
```

```python
sns.displot((y_test-random_preds), bins=100, kde=True)
```
