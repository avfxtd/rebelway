```python
showA = ratings.loc['The Hobbit: An Unexpected Journey']
```

```python
showA
```

```python
showA_sim = ratings.corrwith(showA, axis=1)
```

```python
showA_sim
```

```python
showA_corr = pd.DataFrame(showA_sim, columns=['Correlation'])
showA_corr.dropna(inplace=True)
showA_corr.head()
```

```python
showA_corr.sort_values('Correlation', ascending=False).head()
```

```python
showA_corr = showA_corr.join(ratings['numVotes'])
showA_corr
```

SHOW A RECOMMENDATIONS

```python
showA_corr[showA_corr['numVotes']>5000].sort_values('Correlation', ascending=False).head()
```
