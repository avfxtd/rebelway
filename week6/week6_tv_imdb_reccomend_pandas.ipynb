```python
#ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings = df.set_index('name')[['vote_count','vote_average']]
ratings.head(50)
```

```python
ratings.sort_values('vote_count',ascending=False).head(20)
```

```python
showA = ratings.loc['Squid Game']
```

```python
showA
```

```python
showA_sim = ratings.corrwith(showA, axis=1)
```

```python
showA_sim
```

```python
showA_corr = pd.DataFrame(showA_sim, columns=['Correlation'])
showA_corr.dropna(inplace=True)
showA_corr.head()
```

```python
showA_corr.sort_values('Correlation', ascending=False).head()
```

```python
ratings.head()
```

```python
showA_corr = showA_corr.join(ratings['vote_count'])
showA_corr = showA_corr.join(ratings['vote_average'])
showA_corr
```

SHOW A RECOMMENDATIONS

```python
showA_corr[(showA_corr['vote_count'] > 5000) & (showA_corr['vote_average'] > 2)].sort_values('Correlation', ascending=False).head(10)
```
