```python
plt.bar(df["Pclass"],df["Age"])
```

```python
grouped = df.groupby('Sex')['Age'].value_counts()

for sex, counts in grouped.groupby(level=0):
    plt.figure()
    plt.pie(counts, labels=counts.index.get_level_values(1), autopct='%1.1f%%')
    plt.title(f'Age distribution for {sex}')
    plt.show()
```

```python
# Group by 'Sex' and calculate the mean of 'Survived'
survival_rates = df.groupby('Sex')['Survived'].mean()

# Create a bar chart
plt.bar(survival_rates.index, survival_rates.values)

plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.title('Survival Rates by Sex')
plt.show()
```
